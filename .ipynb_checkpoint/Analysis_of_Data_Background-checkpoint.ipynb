2.1 Market Factor Analysis of Real Estate
---

Draw a chart that combines a histogram and a line graph based on regional distribution. This is done to observe the differences in housing prices among different regions. 

The line graph indicates that the RL region has the highest number of houses and the highest maximum house price. It can be seen that RL belongs to an area with a concentration of residential properties, and the distribution of house prices is relatively flat, following the pyramid rule. However, the FV region has a smaller number of properties, but the average property price is the highest among all regions, indicating that there are fewer houses in FV but they are sold at higher prices, suggesting that it may be an economically developed area. 

Therefore, it can be concluded that the attribute of the property's location is categorical data and has a direct impact on housing prices. Thus, during the data processing stage, this attribute should be considered as a key influencing factor.

```python
# Observe price comparisons in different regions
mean_List = []
max_List = []
min_List = []
region_List = ['RL','RM','C (all)','FV','RH']


for region in region_List:
    
    price_region_df=train_df.loc[train_df['MSZoning']==region]
    price_region_mean = price_region_df['SalePrice'].mean()
    mean_List.append(price_region_mean)

    price_region_max = price_region_df['SalePrice'].max()
    max_List.append(price_region_max)

    price_region_min = price_region_df['SalePrice'].min()
    min_List.append(price_region_min)
    
    

fig, ax = plt.subplots(nrows=1,ncols=1,sharey=True)

# Histogram
sns.set_style('dark')

ax.hist(train_df.MSZoning,bins=10,label = "Region Distribution",color = 'lightsteelblue')

ax2 = ax.twinx()

# Line Chart
ax2.plot(region_List,min_List, 'forestgreen', label='min.plot')
ax2.plot(region_List,max_List, 'red', label='max.plot')
ax2.plot(region_List,mean_List, 'darkorange', label='mean.plot')
ax2.legend()
plt.ylabel("SalePrice")

plt.title("Price Comparisons In Different Regions")
ax.set_xlabel("MSZoning")


plt.show()
```

```python
# A scatter plot of regional distribution and house selling prices
sns.set_style('darkgrid')
plt.scatter(train_df.MSZoning,train_df.SalePrice,color = 'slateblue')
plt.xlabel("MSZoning")
plt.title("MSZoning vs SalePrice")
```

Meanwhile, the sales prices of the properties are compared horizontally and vertically based on the sales year and sales month.

The trend of average property prices in the year-price chart is similar to the trend of average property prices in the month-price chart (represented by the yellow line), both showing a stable and flat slope of 0. This indicates that in the linear relationship between property selling prices and various factors, the correlation coefficient generated by the linear fit between sales year and sales month is 0. Therefore, these two factors can be disregarded.

```python
# Plotting price fluctuations resulting from different sales years
mean_List = []
max_List = []
min_List = []
year_List = [2006,2007,2008,2009,2010]

for year in year_List:
    
    price_year_df=train_df.loc[train_df['YrSold']==year]
    price_year_mean = price_year_df['SalePrice'].mean()
    mean_List.append(price_year_mean)

    price_year_max = price_year_df['SalePrice'].max()
    max_List.append(price_year_max)

    price_year_min = price_year_df['SalePrice'].min()
    min_List.append(price_year_min)
    

fig, ax = plt.subplots(nrows=1,ncols=1,sharey=True)

# Histogram
ax.hist(train_df.YrSold,bins=10,label = "YrSold Distribution",color = 'lightsteelblue')

ax2 = ax.twinx()

# Line Chart
ax2.plot(year_List,min_List, 'forestgreen', label='min.plot')
ax2.plot(year_List,max_List, 'red', label='max.plot')
ax2.plot(year_List,mean_List, 'darkorange', label='mean.plot')
ax2.legend()

plt.title("Price Comparisons In Different Sale Year")
ax.set_xlabel("Year Sold")

plt.show()
```

```python
# Comparing the impact of different months on home sales prices
mean_List = []
max_List = []
min_List = []
month_List = sorted(train_df['MoSold'].unique())


for month in month_List:
    
    price_month_df=train_df.loc[train_df['MoSold']==month]
    price_month_mean = price_month_df['SalePrice'].mean()
    mean_List.append(price_month_mean)

    price_month_max = price_month_df['SalePrice'].max()
    max_List.append(price_month_max)

    price_month_min = price_month_df['SalePrice'].min()
    min_List.append(price_month_min)
    

fig, ax = plt.subplots(nrows=1,ncols=1,sharey=True)

# Histogram
ax.hist(train_df.MoSold,bins=12,label = "MoSold Distribution",color = 'lightsteelblue')

ax2 = ax.twinx()

# Line Chart
ax2.plot(month_List,min_List, 'forestgreen', label='min.plot')
ax2.plot(month_List,max_List, 'red', label='max.plot')
ax2.plot(month_List,mean_List, 'darkorange', label='mean.plot')
ax2.legend()

plt.title("Price Comparisons In Different Sale Month")
ax.set_xlabel("Month Sold")

plt.show()
```

Observing the data, it can be seen that the construction year of properties also has a certain impact on housing prices.

Based on this, a histogram of the distribution of years and a trend graph of property prices every decade were drawn, clearly indicating that the average property price is not significantly correlated with the construction year. However, the maximum selling price of properties shows a clear correlation with the construction year. This indicates that within a certain time period, due to various social factors such as technology and purchasing power, the real estate industry did not aggressively update its products. Therefore, it can be inferred that the housing market prices should be stable within a decade-long time frame.

Specific fluctuations should be attributed more to social factors rather than the inherent quality of the properties themselves.

```python
# The effect of comparing the newness and oldness of the house on the price
mean_List = []
max_List = []
min_List = []
year_List = ['1872-1880','1881-1890','1891-1900','1901-1910','1911-1920','1921-1930','1931-1940','1941-1950','1951-1960','1961-1970',
        '1971-1980','1981-1990','1991-2000','2001-2010']
ax_List = [1875,1885,1895,1905,1915,1925,1935,1945,1955,1965,1975,1985,1995,2005]

for year in year_List:
    min_year=int(year[0:3])
    max_year=int(year[-4:])
    price_year_df=train_df.loc[(min_year<=train_df['YearBuilt'])&(train_df['YearBuilt']<=max_year)]
    price_year_mean = price_year_df['SalePrice'].mean()
    mean_List.append(price_year_mean)

    price_year_max = price_year_df['SalePrice'].max()
    max_List.append(price_year_max)

    price_year_min = price_year_df['SalePrice'].min()
    min_List.append(price_year_min)
    

fig, ax = plt.subplots(nrows=1,ncols=1,sharey=True)

# Histogram
ax.hist(train_df.YearBuilt,bins=10,label = "YearBuilt Distribution",color = 'lightsteelblue')

ax2 = ax.twinx()

# Line Chart
ax2.plot(ax_List,min_List, 'forestgreen', label='min.plot')
ax2.plot(ax_List,max_List, 'red', label='max.plot')
ax2.plot(ax_List,mean_List, 'darkorange', label='mean.plot')
ax2.legend()


plt.title("Mean Price of Each Decade Comparisons In Different Built Year")
ax.set_xlabel("Year Built")


plt.show()
```

According to the histogram of housing prices depicted in the previous text, a new heatmap was generated to illustrate the relationship between the construction year and the property location in terms of housing prices.

Due to the limited number of properties in areas other than RL, the correlation between property prices and construction year was not evident. However, in the RL region, this correlation was validated. Therefore, it is not possible to analyze these two attributes together, but it is feasible to consider merging the location attributes of properties in non-RL areas.

```python
# Heat map of sales prices by year of construction in different regions

Index= ['RL','RM','C (all)','FV','RH']
Cols = ['1872-1880','1881-1890','1891-1900','1901-1910','1911-1920','1921-1930','1931-1940','1941-1950','1951-1960','1961-1970',
        '1971-1980','1981-1990','1991-2000','2001-2010']
df = DataFrame(0, index=Index, columns=Cols)
for i in Index:
    for j in Cols:
        zone_year_df=train_df.loc[train_df['MSZoning']==i]
        min_year=int(j[0:3])
        max_year=int(j[-4:])
        zone_year_df=zone_year_df.loc[(min_year<=zone_year_df['YearBuilt'])&(zone_year_df['YearBuilt']<=max_year)]
        price_mean=zone_year_df['SalePrice'].mean()
        df.loc[[i],[j]]=price_mean
        
plt.title("Heatmap year of YearBuilt and regions")

sns.heatmap(df, annot=False,cmap='OrRd')
```

Contrary to the factors described in the previous context, the heat map generated by the quality and condition of real estate clearly indicates a linear relationship with housing prices. Therefore, these two factors will serve as better parameters for improving the model's fitting effectiveness.

```python
# Price heat map under the influence factors of OverallQual and OverallCond

Index= sorted(train_df.OverallQual.unique(),reverse = True)
Cols = sorted(train_df.OverallCond.unique())

df = DataFrame(0, index=Index, columns=Cols)
for i in Index:
    for j in Cols:
        Qual_Cond_df=train_df.loc[train_df['OverallQual']==i]
        Qual_Cond_df=Qual_Cond_df.loc[train_df['OverallCond']==j]        
        price_mean=Qual_Cond_df['SalePrice'].mean()
        df.loc[[i],[j]]=price_mean

        
sns.heatmap(df, annot=False,cmap='OrRd')

plt.title("Price heatmap under OverallQual and OverallCond")
plt.xlabel("OverallQual")
```

There is no statistical correlation between sales type, sales condition, and the sale price.

However, there is a strong correlation between the "New" sales type and the "Partial" sales condition. This suggests that these two factors are linked under certain conditions in the commercial sales model. Therefore, it is not possible to use these two attributes for linear regression of property sale prices.

```python
# Scatter plots of price under different sales types and scatter plots of sales condition.
sns.set_style('darkgrid')

fig, ax = plt.subplots()
clolrs = train_df['SaleCondition'].unique()
for color in clolrs:
    color_df = train_df.loc[train_df['SaleCondition'] == color]
    ax.scatter(color_df.SaleType, color_df.SalePrice, label=color,
               alpha=0.5, edgecolors='none')
ax.legend()
ax.grid(True)

plt.title("SaleType vs SalePrice")
ax.set_xlabel("SaleType")

plt.show()
```
